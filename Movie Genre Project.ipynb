Movie Genre Data Analysis

Research Questions:
1. Which genres are the most common (number of movies made)?
2. Which genres have high average, budget and revenue?
3. Which genres have the highest average, Profit?   
4. Which genres have  highest average, popularity?
5. Which genres have highest number of movies with an voting average, greater than 8?

Research Hypothesis:

2. The best movies according to popularity return high profit and revenue.
3. Highly budgeted movies returen high revenue and profit.
4. Highly budgeted movies have a high popularity.

```python
import pandas as pd

movies = pd.read_csv(r'C:\Users\charl\OneDrive\Documents\Python Tutorial Notes\imdb_movies.csv')
#movies
```

```python
pd.set_option('display.max.rows', 1100)
pd.set_option('display.max.columns', 25)
```

```python
movies.info()
```

```python
#movies.head()
```

```python
movies.drop_duplicates(inplace = True)
```

```python
# Drop NULL values for id and set the value equal to genres (10842)

movies.dropna(subset = ['genres'], inplace = True)
```

```python
movies.head()
```

```python
movies['profit'] = movies['revenue'] - movies['budget']
```

```python
movies_genres = movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_genres.head()
```

```python
from pandas import Series, DataFrame
```

```python
split = movies_genres['genres'].str.split('|').apply(Series, 1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_genres['genres']
movies_genres = movies_genres.join(split)
```

```python
movies_genres
```

```python
# 1. Which genres are the most common (number of movies made)?
# Answering the first part to the research question, by groupby the genres

movies_genres.groupby('genres_split').count()
```

```python
genres_count = pd.DataFrame(movies_genres.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending = True)
```

```python
genres_count
```

```python
genres_count['original_title'].plot.pie(title = 'Movie Per Genre in %', autopct = '%1.1f%%')
```

```python
genres_count['original_title'].plot.barh(title = 'Movies Per Genre', color = 'DarkGreen')
```

```python
# 2. Which genres have high average budget and revenue?

genres_avg = movies_genres.groupby('genres_split').mean('revenue')
pd.options.display.float_format = '{:2f}'.format

genres_avg
```

```python
genres_avg.sort_values('revenue', ascending = True, inplace = True)
genres_avg
```

```python
genres_avg[['budget','revenue']].plot.barh(title = 'Budget and Revenue by Genre', color = ('DarkBlue', 'c'))
```

```python
# 3.Which genres have the highest average, Profit?

genres_avg.sort_values('profit', ascending = True, inplace = True)
genres_avg['profit'].plot.barh(title = 'Profit by Genre', color = ('Blue'))
```

```python
# 4. Which genres have highest average, popularity?

genres_avg.sort_values('popularity', ascending = True, inplace = True)
genres_avg['popularity'].plot.barh(title = 'Popularity by Genre', color = ('Blue'))
```

```python
#5. Which genres have highest number of movies with an voting average, greater than 8?
```

```python
movies_genres.head()
```

```python
vote_fifty = movies_genres[(movies_genres['vote_count'] >= 50) & (movies_genres['vote_average'] >= 8)]
vote_zero = movies_genres[movies_genres['vote_average'] >= 8]
```

```python
genres_vote = pd.DataFrame(vote_zero.groupby('genres_split').vote_average.nunique()).sort_values('vote_average', ascending = True)
```

```python
genres_vote
```

```python
genres_vote['vote_average'].plot.barh(title = 'Vote Average by Genre', color = 'Purple')
```


Research Hypothesis:

```python
movies.drop_duplicates(inplace = True)
movies['profit'] = movies['revenue'] - movies['budget']
movies_genres = movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
#1. The best movies according to vote average, return the highest profit and revenue.
```

```python
movies_vote = movies_genres[movies_genres['vote_count'] >= 50]
movies_vote[['popularity','budget','revenue','runtime','vote_count','vote_average','profit']].corr(method = 'spearman')
```

```python
import seaborn as sns

sns.regplot(x = 'vote_average', y = 'profit', data = movies_vote, line_kws = {"color": "red"})
```

```python
sns.regplot(x = 'vote_average', y = 'revenue', data = movies_vote, line_kws = {"color": "red"})
```

```python
#2. The best movies according to popularity return high profit and revenue.
movies_vote[['popularity','budget','revenue','runtime','vote_count','vote_average','profit']].corr(method = 'spearman')
```

```python
sns.regplot(x = 'popularity', y = 'revenue', data = movies_vote, line_kws = {"color": "red"})
```

```python
# Bonus question: Look at Profit Per year
```

```python
movies_genres = movies[['popularity','budget','revenue','original_title','runtime','genres','release_year','vote_count','vote_average','profit']]
```

```python
movies_genres.head()
```

```python
time_genre = pd.DataFrame(movies_genres.groupby(['release_year','genres_split'])['profit'].mean())
```

```python
final_genre = pd.pivot_table(time_genre, values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
import matplotlib.pyplot as plt

sns.set(rc = {'figure.figsize':(15,10)})
sns.heatmap(final_genre, cmap = 'YlGnBu', linewidth = .5)
plt.title('Genres by Profit per Year')
```
